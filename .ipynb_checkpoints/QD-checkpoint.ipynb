# ドット中の電子の基底状態
エネルギー固有値は、

$E=0,\pm E_A+\frac{U}{2}$

である。

$E=0$：ダブレット

$E=\pm E_A+\frac{U}{2}$：シングレット

の状態である。ここで、
$E_A=\sqrt{\xi^2_d+|\Delta_d|^2}=\sqrt{\xi^2_d+(\Gamma_R+\Gamma_L)^2\cos^2\frac{\phi}{2}+(\Gamma_R-\Gamma_L)^2\sin^2\frac{\phi}{2}}$である。


```julia
plot(x->x+0.5,xlim=[0,2],label="\$E/U=E_A/U+1/2\$")
plot!(x->-x+0.5,label="\$E/U=-E_A/U+1/2\$")
hline!([0.0],label="\$E/U=0\$");vline!([0.5],color=:red,label="")
xaxis!("\$E_A/U\$",font(15));yaxis!("\$E/U\$",font(15))
annotate!(1.0,-2,text("Singlet"))
annotate!(0.2,-2,text("Doublet"))
```

つまり、基底状態は

$E_A<\frac{U}{2}$はダブレット

$E_A<\frac{U}{2}$はシングレット

の状態をとる。

対称結合$\Gamma_L=\Gamma_R=\Gamma$のとき、

$\frac{\Gamma}{U}<\sqrt{\frac{1}{4\cos^2\frac{\phi}{2}}\left(\frac{1}{4}-(\frac{\xi_d}{U})^2\right)}$:ダブレット

$\frac{\Gamma}{U}>\sqrt{\frac{1}{4\cos^2\frac{\phi}{2}}\left(\frac{1}{4}-(\frac{\xi_d}{U})^2\right)}$:シングレット

```julia
using Plots
g(x,φ)=sqrt(0.25/cos(φ/2)^2*(0.25-x^2))
x=-sqrt(0.25):0.01:sqrt(0.25)
plot(x,g.(x,0.0),ylim=[0.0,0.35],label="\$E_A=\\frac{U}{2}\$")
xaxis!("\$\\xi_d/U\$")
yaxis!("\$\\Gamma/U\$")
annotate!(0,0.3,text("Singlet"))
annotate!(0,0.15,text("Doublet"))
```

```julia
g(x,φ)=sqrt(0.25/cos(φ/2)^2*(0.25-x^2))
x=-sqrt(0.25):0.01:sqrt(0.25)
plot([x],[g.(x,0.0),g.(x,pi/2),g.(x,pi/1.5)],ylim=[0.0,0.7],label=["\$\\phi=0\$","\$\\phi=\\pi/2\$","\$\\phi=\\pi/1.5 \$"])
xaxis!("\$\\xi_d/U\$")
yaxis!("\$\\Gamma/U\$")
annotate!(0,0.6,text("Singlet"))
annotate!(0,0.15,text("Doublet"))
```

# Josephson電流
## Atomic Limit ($\Delta_L,\Delta_R \to \infty$)
$J=\frac{e}{\hbar} \frac{p_- -p_+}{2E_A}4\Gamma_R \Gamma_L \sin \phi$

$p_\pm=\frac{1+e^{-\beta(\frac{U}{2}\pm E_A)}}{2+e^{-\beta(\frac{U}{2}+E_A)}+e^{-\beta (\frac{U}{2}-E_A)}}$

$E_A=\sqrt{\xi_d^2+|\Delta_d|^2}=\sqrt{\xi_d^2+(\Gamma^2_L+\Gamma^2_R)^2-4\Gamma_L\Gamma_R \sin^2\frac{\phi}{2}}$

### $p_- -p_+$のグラフ

```julia
E(ξ,ΓL,ΓR,φ)=sqrt(ξ^2+(ΓR+ΓL)^2 -4*ΓL*ΓR*sin(φ/2)^2 )
Z(T,E)=2+exp(-(1/2-E)/T)+exp(-(1/2+E)/T)
P_m(T,E)=(1+exp(-(1/2-E)/T))/Z(T,E)
P_p(T,E)=(1+exp(-(1/2+E)/T))/Z(T,E)
P(T,E)=P_m(T,E)-P_p(T,E)
e=0.0:0.001:1.5
plot(e ,[P.(0.001,e),P.(0.1,e),P.(0.2,e)],label=["\$T/U=10^{-3}\$","\$T/U=0.1\$","\$T/U=0.2\$"],tickfont=font(11),legend=:bottomright)
xaxis!("\$E_A/U\$");yaxis!("\$P_- -p_+\$")
```

### $\phi$依存
Josephson電流の式において、位相差$\phi$に依存する係数を調べる。

まずは$E_A$の$\phi$依存を調べる。

$E_A=E_{A0}\sqrt{1-\tau\sin^2 \frac{\phi}{2}}$

ここで、

$E_{A0}=\sqrt{\xi_d^2+(\Gamma_L+\Gamma_R)^2}$

$\tau=\frac{4\Gamma_L\Gamma_R}{E_{A0}^2}$

```julia
E_A(x,τ)=sqrt(1-τ*sin(x/2)^2)
φ=0.0:π/100:2*π
plot(φ,[E_A.(φ,0.0),E_A.(φ,0.5),E_A.(φ,1.0)],label=["\$\\tau=0.0\$","\$\\tau=0.5\$","\$\\tau=1.0\$"],legend=:bottomright)
plot!(xticks = ([0:π:2*π;], ["0", "\$\\pi\$", "\$2\\pi\$"]))
xaxis!("\$\\phi\$",font(15));yaxis!("\$E_A/E_{A0}\$")
hline!([0.5],color=:red,label="")
#vline!([2π/3],label="")
```

### $4\Gamma_L\Gamma_R\frac{sin\phi}{E_A}$の$\phi$依存

$4\Gamma_L\Gamma_R\frac{sin\phi}{E_A}=4\Gamma_L\Gamma_R\frac{\sin\phi}{E_{A0}\sqrt{1-T\sin^2 \frac{\phi}{2}}}=E_{A0}\frac{\tau\sin\phi}{\sqrt{1-\tau\sin^2\frac{\phi}{2}}}$

よって、

位相差$\phi$に依存する係数
$\frac{\tau\sin\phi}{\sqrt{1-\tau\sin^2\frac{\phi}{2}}}$
を調べる。

```julia
C_Phi(φ,τ)=τ*sin(φ)*(1-τ*sin(φ/2)^2)
φ=0.0:π/100:2*π
plot(φ,C_Phi.(φ,1.0),label="")
plot!(xticks = ([0:π:2*π;], ["0", "\$\\pi\$", "\$2\\pi\$"]))
xaxis!("\$\\phi\$",font(15))
```

### $p_- -p_+$の$\phi$依存

$E_A=E_{A0}\sqrt{1-\tau \sin^2 \frac{\phi}{2}}$であることから、

$$p_- -p_+=\frac{\exp\left[-\beta\left(\frac{U}{2}- E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}}\right)\right]-\exp[-\beta\left(\frac{U}{2}+ E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}}\right)]}{2+\exp\left[-\beta(\frac{U}{2}+E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}})\right]+\exp\left[-\beta (\frac{U}{2}-E_{A0}\sqrt{1-\tau\sin\frac{\phi}{2}})\right]} $$

```julia
P_phi(φ,τ,T,E)=(exp(-(0.5-E*sqrt(1-τ*sin(0.5*φ)^2))/T)-exp(-(0.5+E*sqrt(1-τ*sin(0.5*φ)^2))/T))/(2+exp(-(0.5-E*sqrt(1-τ*sin(0.5*φ)^2))/T)+exp(-(0.5+E*sqrt(1-τ*sin(0.5*φ)^2))/T))
plot(φ,[P_phi.(φ,1.0,0.001,1.0),P_phi.(φ,1.0,0.1,1.0),P_phi.(φ,1.0,1,1.0)],label=["\$T/U=0.001\$","\$T/U=0.1\$","\$T/U=1\$"])
plot!(xticks = ([0:π:2*π;], ["0", "\$\\pi\$", "\$2\\pi\$"]))
xaxis!("\$\\phi\$",font(15));yaxis!("\$p_- -p_+\$",font(15))
```

位相差$\phi$に依存する全ての因子を考えると、

$\frac{\exp\left[-\beta\left(\frac{U}{2}- E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}}\right)\right]-\exp[-\beta\left(\frac{U}{2}+ E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}}\right)]}{2+\exp\left[-\beta(\frac{U}{2}+E_{A0}\sqrt{1-\tau\sin^2\frac{\phi}{2}})\right]+\exp\left[-\beta (\frac{U}{2}-E_{A0}\sqrt{1-\tau \sin\frac{\phi}{2}})\right]} \frac{E_{A0}\tau\sin\phi}{\sqrt{1-\tau\sin^2\frac{\phi}{2}}}$

```julia
J(φ,τ,T,E)=P_phi(φ,τ,T,E)*C_Phi(φ,τ)
plot(φ,[J.(φ,1.0,0.001,1.0),J.(φ,1.0,0.1,1.0),J.(φ,1.0,1,1.0)],label=["\$T/U=0.001\$","\$T/U=0.1\$","\$T/U=1\$"])
plot!(xticks = ([0:π:2*π;], ["0", "\$\\pi\$", "\$2\\pi\$"]))
xaxis!("\$\\phi\$",font(15));yaxis!("\$J/(e/\\hbar)\$",font(14))
#vline!([2π/3,4π/3],label="")
```
